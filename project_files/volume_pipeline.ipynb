```python
import pandas as pd
from volume_computations import compute_voxel_vol, compute_ch_vol
import os
```

```python
# Define a function to process each row
def process_row(row):
    global vdf
    # set directories
    inf='{a}\{b}\{c}\sor_plant_{c}_v2.ply'.format(a=row['trial_dir'],
                                               b=row['date_dir'],
                                               c=row['plot_id'])
    choutf='{a}\{b}\{c}\ch_{c}'.format(a=row['trial_dir'],
                                       b=row['date_dir'],
                                       c=row['plot_id'])
    
    # Run voxel volume function
    w, d, h, v = compute_voxel_vol(inf)
    print(row['sample_id'] + ' voxel volume calculated')

    # Run convex hull volume function
    cv, ga = compute_ch_vol(inf, choutf)
    print(row['sample_id'] + ' convex hull volume calculated')

    # Add the new row to the volume directory
    vrow = pd.DataFrame({'sample_id':row['sample_id'],
                         'plot_id':row['plot_id'],
                         'trial_dir':row['trial_dir'],
                         'date_dir':row['date_dir'],
                         'species':row['species'],
                         'width':w,
                         'depth':d,
                         'height':h,
                         'voxel_vol_02':v[0],
                         'voxel_vol_01':v[1],
                         'voxel_vol_005':v[2],
                         'voxel_vol_002':v[3],
                         'voxel_vol_001':v[4],
                         'convh_vol':cv,
                         'ground_area':ga}, index=[0])
    vdf=pd.concat([vdf, vrow], axis=0, ignore_index=True)
    print(row['sample_id'] + ' added to parameter dataframe')

    return


```

```python
# Read in df
df = pd.read_csv('full_pc_directory.csv')
```

```python
# Loop for cucumber fertilizer trial

vcols = ['sample_id', 'plot_id', 'trial_dir', 'date_dir', 'species', 'width', 'depth', 'height',
         'voxel_vol_02', 'voxel_vol_01', 'voxel_vol_005', 'voxel_vol_002', 'voxel_vol_001', 'convh_vol', 'ground_area']
vdf = pd.DataFrame(columns=vcols)
filename= 'volume_directory.csv'

dfc = df.loc[(df['trial_dir'] == 'Cucumber_Fertilizer_Trial')]

dfc.apply(process_row, axis=1)

# Check if the file already exists
if os.path.isfile(filename):
    # Append to the existing file
    vdf.to_csv(filename, mode='a', header=False, index=False)
else:
    # Create a new file
    vdf.to_csv(filename, index=False)
```
