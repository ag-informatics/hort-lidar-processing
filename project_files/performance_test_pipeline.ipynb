```python
import pandas as pd
import numpy as np
import open3d as o3d
import gc
from perf_tests import merge_nonplant_pc, merge_nonplant_subsample_pc, iou
import os
```

```python
# Loop for entire df

# Load directory df
df_total = pd.read_csv('full_pc_directory.csv')
df_total.set_index('sample_id', inplace=True, drop=False)

# Subset df to only include performance test samples
df = df_total.loc[df_total['s_priority'] != 5]

cols = ['sample','groundtruthed_total', 'processed_total','groundtruthed_plant_total',
        'groundtruthed_nonplant_total','processed_plant_total','processed_nonplant_total',
        'ppgpint','pngnint','ppgnint','pngpint','punion','nunion','piou','niou']

df_iou = pd.DataFrame(columns=cols)

def process_row(row):
    global df_iou
    trial_dir = row['trial_dir']
    date_dir = row['date_dir']
    plot_id = row['plot_id']
    # Merge nonplant point clouds
    merge_nonplant_pc(itrial_dir=trial_dir,
                      date_dir=date_dir,
                      plot_id=plot_id,
                      otrial_dir=trial_dir)
    print('Merged nonplant point clouds for {a}'.format(a=row['sample_id']))
    # Merge nonplant subsample point clouds
    merge_nonplant_subsample_pc(itrial_dir=trial_dir,
                                date_dir=date_dir,
                                plot_id=plot_id,
                                otrial_dir=trial_dir)
    print('Merged nonplant subsample point clouds for {a}'.format(a=row['sample_id']))
    # Conduct intersection over union tests
    df_row=iou(sample=row['sample_id'],
                               gtrial_dir=trial_dir,
                               ptrial_dir=trial_dir,
                               date_dir=date_dir,
                               plot_id=plot_id)
    df_iou=pd.concat([df_iou, df_row], axis=0, ignore_index=True)
    print('Completed intersection over union tests for {a}'.format(a=row['sample_id']))
    return

df.apply(process_row, axis=1)

df_iou.to_csv('performance_test_iou.csv', index=False)
```
