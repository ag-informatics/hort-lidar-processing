```python
import pandas as pd
from pfunctions import las_to_ply, convert_colors, c_thresh_seg, crop_b_box, sor_outlier_removal, crop_by_pc
```

```python
# Define a function to process each row
def process_row(row):
    # set directories
    trial_dir = row['trial_dir']
    date_dir = row['date_dir']
    plot_id = row['plot_id']
    # run example las_to_ply
    las_to_ply(itrial_dir=trial_dir,
                date_dir=date_dir,
                plot_id=plot_id, 
                otrial_dir=trial_dir)
    print('{z} {a} {b} conversion to .ply complete'.format(z=trial_dir, a=date_dir,b=plot_id))
    # generate an HSV color space ply
    convert_colors(itrial_dir=trial_dir,
                   date_dir=date_dir,
                   plot_id=plot_id, 
                   otrial_dir=trial_dir)
    print('{z} {a} {b} hsv coloration complete'.format(z=trial_dir, a=date_dir,b=plot_id))
    # use color thresholding to segment plant from non-plant points
    c_thresh_seg(itrial_dir=trial_dir,
                 date_dir=date_dir,
                 plot_id=plot_id, 
                 otrial_dir=trial_dir,
                 h=row['hue'],
                 s=row['saturation'],
                 v =row['value'])
    print('{z} {a} {b} segmentation by color complete'.format(z=trial_dir, a=date_dir,b=plot_id))
    # crop plant by plot boundaries
    crop_b_box(itrial_dir=trial_dir,
               date_dir=date_dir,
               plot_id=plot_id, 
               otrial_dir=trial_dir,
               width=row['bounding_width'],
               depth=row['bounding_depth'],
               height=row['bounding_height'])
    print('{z} {a} {b} cropping by plot boundaries complete'.format(z=trial_dir, a=date_dir,b=plot_id))
    # remove outliers with sor
    sor_outlier_removal(itrial_dir=trial_dir,
                        date_dir=date_dir,
                        plot_id=plot_id, 
                        otrial_dir=trial_dir,
                        k=row['k_s'],
                        std_ratio=row['s_std_ratio'])
    print('{z} {a} {b} outlier removal by sor complete'.format(z=trial_dir, a=date_dir,b=plot_id))
        # crop raw point cloud by cropped bounding box point cloud for later annotation
    crop_by_pc(rtrial_dir=trial_dir,
               date_dir=date_dir,
               plot_id=plot_id, 
               ttrial_dir=trial_dir)
    print('{z} {a} {b} cropping raw point cloud by cropped bounding box point cloud complete'.format(z=row['trial_dir'], a=date_dir,b=plot_id))



```

```python
# Load directory df
df = pd.read_csv('full_pc_directory.csv')
```

```python
# Loop for garden plots
dfg = df.loc[(df['trial_dir'] == 'Garden_Plots')]
dfg.apply(process_row, axis=1)
```
